```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# Datos

1) Llamar la base de datos

2) Seleccionar las columnas que se utilizarán para los gráficos de disperción

3) Convertir las columnas a listas para hacer las conversiones de masa y facilitar gráficas

```python
#NASA-----> [MA]=[UA]
#---------> [mass]=[Mjup]
conv= 317.83
NASA=pd.read_csv('NASAplanets.csv', delimiter=',')
#-------------------------------------------
MajorAxisNAS=NASA.iloc[:,6:7]
MajorAxisNAS=MajorAxisNAS.values.tolist()
MajorAxisNASA=[]
for i in range(len(MajorAxisNAS)):
    MajorAxisNASA+=MajorAxisNAS[i]  
#-------------------------------------------    
MassNAS=NASA.iloc[:,9:10]
MassNAS=MassNAS.values.tolist()
MassNASA=[]
for i in range (len(MassNAS)):
    MassNASA+=MassNAS[i]
MassNASAEarth=[]
MassNASAEarth=[i*conv for i in MassNASA]
```

```python
#ExoplanetsEU-----> [MA]=[UA]
#        ---------> [mass]=[Mjup]
exo=pd.read_csv('exoplaneteu_catalog.csv', delimiter=',')
#--------------------------------------
MajorAxisEX=exo.iloc[:,14:15]
MajorAxisEX=MajorAxisEX.values.tolist()
MajorAxisEXO=[]
for i in range(len(MajorAxisEX)):
    MajorAxisEXO+=MajorAxisEX[i]
#--------------------------------------   
MassEX=exo.iloc[:,2:3]
MassEX=MassEX.values.tolist()
MassEXO=[]
for i in range(len(MassEX)):
    MassEXO+=MassEX[i]
MassEXOEarth=[i*conv for i in MassEXO]
```

```python
#Exoplanets-----> [MA]=[UA]     #HUBO DATOS FALTANTES EN EJE MAYOR, DICHOS DATOS FUERON COMPLETADOS CON nan
#        ---------> [mass]=[Mjup]               ADEMÁS LA BASE DE DATOS NO PERMITIÓ LLAMAR ESA COLUMNA
expo=pd.read_csv('Baseexoplanetsorg.csv', delimiter=',')

#--------------------------------------
MajorAxisEXP=np.loadtxt('baseaxis.txt')
MajorAxisEXP=MajorAxisEXP.tolist()      
#--------------------------------------   
MassEXPo=expo.iloc[1:,2:3]
MassEXPo=MassEXPo.values.tolist()
MassEXPO=[]
for j in range (len(MassEXPo)):
    MassEXPO+=MassEXPo[j]
MassEXPO=[float(i) for i in MassEXPO]    
MassEXPOEarth=[]
for elem in MassEXPO:
    MassEXPOEarth.append(elem*conv)    
```

# Gráficas de dispersión para masas Jupiter



```python
#NASA
scaNASA_J=plt.scatter(x=MajorAxisNASA, y=MassNASA, marker='o',s=20, color='brown', alpha=0.8, label='NASA')
plt.xscale('log')
plt.xlim(0.005,1000)
plt.title('Gráfico de dispersión para datos NASA')
plt.grid()
plt.xlabel('Semieje Mayor (UA)')
plt.ylabel('Masa (M♃)')
#plt.savefig('name.pdf')
plt.show()
```

```python
#ExoplanetsEU
scaEXOEU_J=plt.scatter(x=MajorAxisEXO, y=MassEXO, marker='o', color='green',s=20, alpha=0.8, label='ExoplanetEU')
plt.xscale('log')
plt.xlim(0.005,1000)
plt.title('Gráfico de dispersión para datos ExoplanetEU')
plt.grid()
plt.xlabel('Semieje Mayor (UA)')
plt.ylabel('Masa (M♃)')
#plt.savefig('name.pdf')
plt.show()
```

```python
#Exoplanets
scaEXO_J=plt.scatter(x=MajorAxisEXP, y=MassEXPO, marker='o', color='blue',s=20, alpha=0.8, label='Exoplanet')
plt.xscale('log')
plt.xlim(0.005,1000)
plt.title('Gráfico de dispersión para datos Exoplanet')
plt.grid()
plt.xlabel('Semieje Mayor (UA)')
plt.ylabel('Masa (M♃)')
#plt.savefig('name.pdf')
plt.show()
```
