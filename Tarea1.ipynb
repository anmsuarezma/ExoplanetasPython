```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# Datos

1) Llamar la base de datos

2) Seleccionar las columnas que se utilizarán para los gráficos de disperción

3) Convertir las columnas a listas para hacer las conversiones de masa y facilitar gráficas

```python
#NASA-----> [MA]=[UA]
#---------> [mass]=[Mjup]
conv= 317.83
NASA=pd.read_csv('BaseexoplanetsNASAfil.csv', delimiter=',', encoding='latin-1')
#-------------------------------------------
MajorAxisNAS=NASA.iloc[:,6:7]
MajorAxisNAS=MajorAxisNAS.values.tolist()
MajorAxisNASA=[]
for i in range(len(MajorAxisNAS)):
    MajorAxisNASA+=MajorAxisNAS[i]  
#-------------------------------------------    
MassNAS=NASA.iloc[:,9:10]
MassNAS=MassNAS.values.tolist()
MassNASA=[]
for i in range (len(MassNAS)):
    MassNASA+=MassNAS[i]
MassNASAEarth=[]
MassNASAEarth=[i*conv for i in MassNASA]
```

```python
#ExoplanetsEU-----> [MA]=[UA]
#        ---------> [mass]=[Mjup]
exo=pd.read_csv('BaseexoplanetsESAfil.csv', delimiter=',', encoding='latin-1')
#--------------------------------------
MajorAxisEX=exo.iloc[:,4:5]
MajorAxisEX=MajorAxisEX.values.tolist()
MajorAxisEXO=[]
for i in range(len(MajorAxisEX)):
    MajorAxisEXO+=MajorAxisEX[i]
#--------------------------------------   
MassEX=exo.iloc[:,1:2]
MassEX=MassEX.values.tolist()
MassEXO=[]
for i in range(len(MassEX)):
    MassEXO+=MassEX[i]
MassEXOEarth=[i*conv for i in MassEXO]
```

```python
#Exoplanets-----> [MA]=[UA]     #HUBO DATOS FALTANTES EN EJE MAYOR, DICHOS DATOS FUERON COMPLETADOS CON nan
#        ---------> [mass]=[Mjup]               ADEMÁS LA BASE DE DATOS NO PERMITIÓ LLAMAR ESA COLUMNA
expo=pd.read_csv('BaseexoplanetsORGfil.csv', delimiter=',', encoding='latin-1')
MajorAxisEXPO=expo.iloc[:,1:2]
MajorAxisEXPO=MajorAxisEXPO.values.tolist()
MajorAxisEXP=[]
for i in range(len(MajorAxisEXPO)):
    MajorAxisEXP+=MajorAxisEXPO[i]
MajorAxisEXP=[float(j) for j in  MajorAxisEXP]    
#--------------------------------------   
MassEXP=expo.iloc[:,2:3]
MassEXP=MassEXP.values.tolist()
MassEXPO=[]
for i in range(len(MassEXP)):
    MassEXPO+=MassEXP[i]
MassEXPO=[float (j) for j in MassEXPO]
MassEXPOEarth=[j*conv for j in MassEXPO]  
MajorAxisEXP
```

# Gráficas de dispersión para masas Jupiter



```python
#NASA
scaNASA_J=plt.scatter(x=MajorAxisNASA, y=MassNASA, marker='o',s=20, color='brown', alpha=0.8, label='NASA',zorder=2)

#plt.figure()
plt.xscale('log')
plt.xlim(0.005,1000)
plt.title('Gráfico de dispersión para datos NASA')
plt.xlabel('Semieje Mayor (UA)')
plt.ylabel('Masa (M♃)')
#plt.savefig('name.pdf')
plt.grid(zorder=1, linestyle='--')
plt.show()
```

```python
#ExoplanetsEU
scaEXOEU_J=plt.scatter(x=MajorAxisEXO, y=MassEXO, marker='o', color='green',s=20, alpha=1.0, label='ExoplanetEU',zorder=2)
plt.xscale('log')
plt.xlim(0.005,1000)
plt.title('Gráfico de dispersión para datos ExoplanetEU')
plt.grid(zorder=1, linestyle='--')
plt.xlabel('Semieje Mayor (UA)')
plt.ylabel('Masa (M♃)')
plt.savefig('name1.pdf')
plt.show()
```

```python
#Exoplanets
scaEXO_J=plt.scatter(x=MajorAxisEXP, y=MassEXPO, marker='o', color='blue',s=20, alpha=0.8, label='Exoplanet',zorder=2)
plt.xscale('log')
plt.xlim(0.005,1000)
plt.title('Gráfico de dispersión para datos Exoplanet')
plt.grid(zorder=1, linestyle='--')
plt.xlabel('Semieje Mayor (UA)')
plt.ylabel('Masa (M♃)')
#plt.savefig('name.pdf')
plt.show()
```
